```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.ensemble import RandomForestRegressor
```

```python
#load in the data using the pandas library and display first few data sets. 
df = pd.read_csv('Breast_cancer_data.csv')
df.head()
```

```python
#seperate the data in x - the independents, y - the dependents(result)
y = df['diagnosis']
y
```

```python
#we get all the x values by removing all the y values 
x = df.drop('diagnosis',axis=1)
x
```

```python
#spit the data set into training and testing set
x_train,x_test,y_train,y_test = train_test_split(x, y, test_size=0.2, random_state=100)
```

```python
x_train
```

```python
x_test
```

```python
#Model(linear Regression) -Training

lr = LinearRegression()
#to train linear regression the x_train and y_train data set
lr.fit(x_train,y_train)
```

```python
#Make a prediction.
y_lr_train_pred = lr.predict(x_train)
y_lr_test_pred = lr.predict(x_test)
```

```python
#Prediction for the train x_variables
y_lr_train_pred
```

```python
#Prediction for the test x_variables
y_lr_test_pred
```

```python
#To evaluate model performance 

lr_train_mse = mean_squared_error(y_train, y_lr_train_pred)
lr_train_r2 = r2_score(y_train, y_lr_train_pred)

lr_test_mse = mean_squared_error(y_test, y_lr_test_pred)
lr_train_r2 = r2_score(y_test, y_lr_test_pred)

#to put the results in data frame
lr_results = pd.DataFrame(['Linear Regression', lr_train_mse, lr_train_r2, lr_test_mse, lr_train_r2]).transpose()
lr_results.columns = ['Method', 'Training(MSE)', 'Training(r2s)', 'Test(MSE)', 'Test(r2s)']

lr_results
```

```python
#Model(Random Forest) -Training

rf = RandomForestRegressor(max_depth=2, random_state=100)
rf.fit(x_train, y_train)
```

```python
#Make a prediction.
y_rf_train_pred = rf.predict(x_train)
y_rf_test_pred = rf.predict(x_test)
```

```python
#To evaluate model performance 

rf_train_mse = mean_squared_error(y_train, y_rf_train_pred)
rf_train_r2 = r2_score(y_train, y_rf_train_pred)

rf_test_mse = mean_squared_error(y_test, y_rf_test_pred)
rf_train_r2 = r2_score(y_test, y_rf_test_pred)

#to put the results in data frame
rf_results = pd.DataFrame(['Random Forest Regression', rf_train_mse, rf_train_r2, rf_test_mse, rf_train_r2]).transpose()
rf_results.columns = ['Method', 'Training(MSE)', 'Training(r2s)', 'Test(MSE)', 'Test(r2s)']

rf_results
```
